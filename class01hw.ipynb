## 卷积核

```python
import cv2 as cv

Einstein = cv.imread('Einstein.jpg')

plt.imshow(Einstein)

kernel = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]])

edges = cv.filter2D(Einstein, -1, kernel)
cv.imwrite('Einstein-edges.jpg', edges)
```

```python
plt.imshow(edges)
```
